```python
# Install libraries (run once)
!pip install transformers torch

# Import pipeline
from transformers import pipeline

# Load Question Answering model
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

print("Chatbot is ready! Type 'exit' to stop.\n")

# Context (knowledge base)
context = """
Artificial Intelligence is the simulation of human intelligence in machines.
It allows systems to learn, reason, and solve problems.

Python is a programming language created by Guido van Rossum in 1991.
It is widely used in web development, data science, and machine learning.

Machine learning is a subset of Artificial Intelligence that allows systems to learn from data.

Cloud computing is the delivery of computing services like storage and servers over the internet.
"""

# Chat loop
while True:
    user_input = input("You: ")

    # Exit condition
    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot: Goodbye!")
        break

    # Greeting handling
    elif user_input.lower() in ["hello", "hi", "hey"]:
        print("Chatbot: Hello! How can I help you today?")

    # Question answering
    else:
        result = qa_pipeline(
            question=user_input,
            context=context
        )

        print("Chatbot:", result['answer'])
```
