### Task 1: Conceptual Understanding

### 1. Difference between "Love" and "love" in NLP

#####  In NLP, "Love" and "love" are considered different words because the system treats uppercase and lowercase letters separately. This is called case sensitivity. To avoid treating them as different, we usually convert all text into lowercase so both become the same word "love".

### 2. What happens if stopwords are not removed?

##### If stopwords are not removed, the text will contain many unnecessary words like "is", "the", "and". These words do not add much meaning but increase the size of the data. This can make the model slower and less accurate because important words get less focus.

### 3. Two real-world scenarios where removing stopwords can be harmful

### 1. Sentiment Analysis
##### In sentences like "I am not happy", the word "not" is very important. If we remove it, the meaning changes completely to "I am happy", which is incorrect.

### 2. Question Answering Systems
##### In questions like "What is your name?", words like "what" and "is" are important to understand the question. Removing them can confuse the system and give wrong answers.

### 4. Difference between stemming and lemmatization

##### Stemming is a process that cuts words to their root form, but the result may not be a real word. For example, "running" becomes "runn".
##### Lemmatization is more advanced and converts words into their correct base form using meaning. For example, "running" becomes "run".
#### So, lemmatization gives more accurate and meaningful results compared to stemming.







### Task 2:Build Advanced Preprocessing Function

```python
import re
import string

def preprocess_text(text):
    # Error handling
    if not text or not isinstance(text, str):
        return []

    # Remove URLs & emails
    text = re.sub(r'http\S+|www\S+|\S+@\S+', '', text)

    # Remove emojis
    text = re.sub(r'[^\x00-\x7F]+', '', text)

    # Convert to lowercase
    text = text.lower()

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Handle repeated characters (soooo → soo)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # Tokenization
    tokens = text.split()

    # Remove short words (except "no", "not")
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    return tokens


# ==============================
# TEST (VERY IMPORTANT)
# ==============================

sample_text = "I absolutely looooved this product 😍😍 100%!!! Visit https://abc.com"

result = preprocess_text(sample_text)

print("Original Text:", sample_text)
print("Processed Tokens:", result)
print("Clean Sentence:", " ".join(result))
```

### Task 3: Stress Testing


```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for text in test_sentences:
    tokens = preprocess_text(text)
    print("\nOriginal:", text)
    print("Tokens:", tokens)
    print("Clean Sentence:", " ".join(tokens))
```

### Task 4: Token Analytics

```python
def token_analysis(tokens):
    total = len(tokens)
    unique = len(set(tokens))
    avg = sum(len(w) for w in tokens)/total if total > 0 else 0
    return total, unique, avg

for text in test_sentences:
    tokens = preprocess_text(text)
    total, unique, avg = token_analysis(tokens)

    print("\nText:", text)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Length:", avg)
```

### Analysis Answers

- Sentence with most noise: "Get 100% FREE access now !"
- Sentence with most meaningful tokens: "I am not happy with this"

### Task 5: Frequency Analysis

```python
from collections import Counter

all_tokens = []

for text in test_sentences:
    all_tokens.extend(preprocess_text(text))

counter = Counter(all_tokens)

print("Top 10 frequent:", counter.most_common(10))
print("Least 5 frequent:", counter.most_common()[-5:])
```

### Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(" ".join(tokens))

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

result = full_pipeline(test_sentences)
print(result)
```
